```python
import matplotlib.pyplot as plt

# Sample data: Ratings (on a scale from 1 to 5, for example)
ratings = data['Rating']

# Function to generate a stem-and-leaf plot for ratings
def stem_and_leaf_plot(data):
    # Sort the data first
    data_sorted = sorted(data)

    # Create stems (integer part) and leaves (decimal part)
    stems = [int(n) for n in data_sorted]  # Integer part (stem)
    leaves = [round((n - int(n)) * 10) for n in data_sorted]  # Decimal part * 10 (leaf)

    # Grouping by stems
    stem_dict = {}
    for stem, leaf in zip(stems, leaves):
        if stem in stem_dict:
            stem_dict[stem].append(leaf)
        else:
            stem_dict[stem] = [leaf]

    # Print the stem-and-leaf plot
    print("Stem | Leaf")
    print("-------------")
    for stem, leaf_list in stem_dict.items():
        leaf_str = ' '.join(map(str, sorted(leaf_list)))
        print(f"  {stem}  | {leaf_str}")

# Call the function with the sample ratings data
stem_and_leaf_plot(ratings)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
import time

# Load the supermarket dataset
data = pd.read_csv('/content/supermarket_sales - Sheet1.csv')  # Adjust path if needed

# Preprocessing: Create a basket for each transaction (Invoice ID) with items (Product line)
basket = (data.groupby(['Invoice ID', 'Product line'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('Invoice ID'))

# Convert quantities to binary (0 or 1), required for Apriori
basket = basket.applymap(lambda x: 1 if x > 0 else 0)

def run_apriori(data, min_support, min_confidence):
    print(f"\nRunning Apriori with Support={min_support}, Confidence={min_confidence}")
    start_time = time.time()

    # Find frequent itemsets using the Apriori algorithm
    frequent_itemsets = apriori(data, min_support=min_support, use_colnames=True)

    # Generate association rules
    if not frequent_itemsets.empty:
        rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence,num_itemsets=0)
        end_time = time.time()
        exec_time = end_time - start_time
        num_rules = len(rules)
        print(f"Execution Time: {exec_time:.4f} seconds")
        return exec_time, num_rules, rules
    else:
        print("No frequent itemsets found.")
        return 0, 0, pd.DataFrame()

# Define support and confidence values to test
support_values = [0.001, 0.002, 0.003]
confidence_values = [0.4, 0.5, 0.6]

# Run Apriori with different parameter combinations
times = []
rules_counts = []
labels = []
all_rules = []

for support in support_values:
    for confidence in confidence_values:
        exec_time, num_rules, rules = run_apriori(basket, min_support=support, min_confidence=confidence)
        times.append(exec_time)
        rules_counts.append(num_rules)
        labels.append(f"S={support}, C={confidence}")
        all_rules.append(rules)  # Store rules for further analysis if needed

# Plotting Execution Time
plt.figure(figsize=(10, 5))
plt.bar(labels, times, color='skyblue')
plt.xticks(rotation=45)
plt.xlabel('Support-Confidence Pairs')
plt.ylabel('Execution Time (seconds)')
plt.title('Apriori Algorithm Execution Time for Different Parameters')
plt.show()

# Optional: Display first few rules for review
for i, rules in enumerate(all_rules):
    if not rules.empty:
        print(f"\nRules for {labels[i]}:")
        display(rules.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.preprocessing import TransactionEncoder
import time

# Load the supermarket dataset
data = pd.read_csv('/content/supermarket_sales - Sheet1.csv')  # Adjust path if needed

# Preprocessing: Create transactions by grouping by Invoice ID and Product Line
transactions = data.groupby('Invoice ID')['Product line'].apply(list).values.tolist()

# Transaction Encoding
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

# Define the FP-Growth function
def run_fp_growth(data, min_support):
    print(f"\nRunning FP-Growth with Support={min_support}")
    start_time = time.time()

    # Find frequent itemsets using FP-Growth algorithm
    frequent_itemsets = fpgrowth(data, min_support=min_support, use_colnames=True)

    # Execution time
    exec_time = time.time() - start_time
    num_itemsets = len(frequent_itemsets)

    # Output
    print(f"Execution Time: {exec_time:.4f} seconds")
    print(f"Number of Frequent Itemsets: {num_itemsets}")

    return exec_time, num_itemsets

# Define support values to test
support_values = [0.001, 0.002, 0.003]  # Lower support values to capture more itemsets
execution_times = []
itemset_counts = []

# Run FP-Growth for different support values
for support in support_values:
    exec_time, num_itemsets = run_fp_growth(df, min_support=support)
    execution_times.append(exec_time)
    itemset_counts.append(num_itemsets)

# Plotting Execution Time for different support values
plt.figure(figsize=(10, 5))
plt.bar([f"Support={s}" for s in support_values], execution_times, color='skyblue')
plt.xlabel('Support Values')
plt.ylabel('Execution Time (seconds)')
plt.title('FP-Growth Execution Time for Different Support Values')
plt.show()

# Plotting Number of Frequent Itemsets for different support values
plt.figure(figsize=(10, 5))
plt.bar([f"Support={s}" for s in support_values], itemset_counts, color='lightgreen')
plt.xlabel('Support Values')
plt.ylabel('Number of Frequent Itemsets')
plt.title('Number of Frequent Itemsets for Different Support Values')
plt.show()
```

```python
# DIC
import pandas as pd
from itertools import combinations
from collections import defaultdict
import psutil  # For profiling memory usage

class DIC:
    def __init__(self, min_support, max_itemset_size=3):
        self.min_support = min_support
        self.max_itemset_size = max_itemset_size
        # Initialize the dictionaries
        self.item_count = defaultdict(int)
        self.transaction_list = []
        self.itemsets = defaultdict(int)

    def load_data(self, file_path):
        """Load dataset and extract product lines per invoice."""
        print("Loading data...")
        df = pd.read_csv(file_path)  # Load CSV properly
        # Group by 'Invoice ID' and get the 'Product line' values into lists
        transactions = df.groupby('Invoice ID')['Product line'].apply(list).values.tolist()
        return transactions

    def first_pass(self, transactions):
        """Count single item occurrences."""
        print("First pass: counting individual items...")
        for transaction in transactions:
            for item in transaction:
                self.item_count[item] += 1
        # Keep only items that meet the minimum support
        self.item_count = {item: count for item, count in self.item_count.items() if count >= self.min_support}
        # Create transaction list of only frequent items
        self.transaction_list = [
            [item for item in transaction if item in self.item_count]
            for transaction in transactions
        ]

    def dynamic_counting(self):
        """Dynamic counting of itemsets."""
        print(f"Generating itemsets up to size {self.max_itemset_size}...")
        for transaction in self.transaction_list:
            # Create all combinations of items in the transaction
            for r in range(2, min(self.max_itemset_size + 1, len(transaction) + 1)):
                for itemset in combinations(transaction, r):
                    # Convert all items to strings before sorting to avoid type errors
                    sorted_itemset = tuple(sorted(itemset, key=str))
                    self.itemsets[sorted_itemset] += 1

    def get_frequent_itemsets(self):
        """Filter itemsets based on min_support."""
        return {itemset: count for itemset, count in self.itemsets.items() if count >= self.min_support}

    def run(self, file_path):
        """Run the DIC algorithm to find frequent itemsets."""
        # Monitor memory usage before loading data
        print(f"Memory usage before loading data: {psutil.virtual_memory().percent}%")
        # Load and process data
        transactions = self.load_data(file_path)
        print(f"Memory usage after loading data: {psutil.virtual_memory().percent}%")
        # First pass: single item frequency counting
        self.first_pass(transactions)
        print(f"Memory usage after first pass: {psutil.virtual_memory().percent}%")
        # Second pass: dynamic counting of itemsets
        self.dynamic_counting()
        print(f"Memory usage after dynamic counting: {psutil.virtual_memory().percent}%")
        # Return the frequent itemsets that meet the minimum support
        return self.get_frequent_itemsets()

def main():
    # Define minimum support and maximum itemset size
    min_support = 100  # Adjust based on your dataset size
    max_itemset_size = 3  # Set a reasonable max size to control memory usage
    # Run the DIC algorithm
    dic = DIC(min_support, max_itemset_size)
    file_path = '/content/supermarket_sales - Sheet1.csv'  # Replace with your path to the transaction dataset
    frequent_itemsets = dic.run(file_path)
    # Display results
    print("Frequent Itemsets:")
    if frequent_itemsets:
        for itemset, count in frequent_itemsets.items():
            print(f"Itemset: {itemset}, Count: {count}")
    else:
        print("No frequent itemsets found.")

if __name__ == "__main__":
    main()
```

```python
# transaction reduction
# apriori variant
import pandas as pd
from mlxtend.frequent_patterns import apriori, fpgrowth
def load_data():
  dat_file_path = '/content/supermarket_sales - Sheet1.csv'
  transactions = []
  with open(dat_file_path, 'r') as file:
    for line in file:
      transaction = line.strip().split()
      transactions.append(transaction)
  data = pd.DataFrame(transactions)
  return data
def preprocess_data(data):
    return pd.get_dummies(data.stack()).groupby(level=0).sum()

def apriori_with_transaction_reduction(data, min_support=0.3):
  frequent_items = apriori(data, min_support=min_support, use_colnames=True)
  frequent_item_set = set().union(*frequent_items['itemsets'])
  reduced_data = data[data.columns.intersection(frequent_item_set)]
  frequent_itemsets_reduced = apriori(reduced_data, min_support=min_support, use_colnames=True)
  return frequent_itemsets_reduced
def fp_growth(data, min_support=0.3):
  return fpgrowth(data, min_support=min_support, use_colnames=True)
data = load_data()
encoded_data = preprocess_data(data)
apriori_results = apriori_with_transaction_reduction(encoded_data, min_support=0.3)
print("Apriori with Transaction Reduction Results:")
print(apriori_results)
fp_growth_results = fp_growth(encoded_data, min_support=0.3)
print("\nFP-Growth Results:")
print(fp_growth_results)
```

```python
# vertical transormation
import pandas as pd
from collections import defaultdict

def load_mushroom_data(file_path):
    transactions = []
    with open(file_path, 'r') as file:
        for line in file:
            items = line.strip().split()
            transactions.append(items)
    return transactions

def vertical_format(transactions):
    vertical_db = defaultdict(list)
    for tid, transaction in enumerate(transactions):
        for item in transaction:
            vertical_db[item].append(tid)
    return vertical_db

def calculate_support(vertical_db, min_support):
    support_count = {}
    for item, tids in vertical_db.items():
        if len(tids) >= min_support:
            support_count[item] = tids
    return support_count

def build_fp_tree(vertical_db, min_support):

    support_count = calculate_support(vertical_db, min_support)


    sorted_items = sorted(support_count.items(), key=lambda x: len(x[1]), reverse=True)

    fp_tree = {}
    for item, tids in sorted_items:
        fp_tree[item] = tids

    return fp_tree

def main():

    file_path = '/content/supermarket_sales - Sheet1.csv'
    transactions = load_mushroom_data(file_path)

    vertical_db = vertical_format(transactions)


    min_support = 100


    fp_tree = build_fp_tree(vertical_db, min_support)


    print("FP-tree Itemsets with Transaction IDs:")
    for item, tids in fp_tree.items():
        print(f"Item: {item}, Transactions: {tids}")

if __name__ == "__main__":
    main()
```

```python
#eclat
import pandas as pd
from itertools import combinations

# Step 1: Load the dataset
dataset_path = '/content/supermarket_sales - Sheet1.csv'  # Replace with your dataset's file path
df = pd.read_csv(dataset_path)

# Step 2: Prepare the transactional dataset
# Group data by transactions (e.g., InvoiceNo) and aggregate items
transaction_col = 'Invoice ID'  # Transaction ID column
item_col = 'Product line'  # Item column
transactions = df.groupby(transaction_col)[item_col].apply(set)

# Convert to a dictionary for easier processing
transaction_dict = transactions.to_dict()

# Step 3: Define ECLAT functions
def get_support(itemset, transaction_dict):
    """Calculate support for an itemset."""
    support = sum(1 for transaction in transaction_dict.values() if itemset.issubset(transaction))
    return support

def eclat(prefix, items, min_support, transaction_dict, frequent_itemsets):
    """Recursive ECLAT algorithm to find frequent itemsets."""
    while items:
        item = items.pop()
        itemset = prefix.union({item})
        support = get_support(itemset, transaction_dict)

        if support >= min_support:
            frequent_itemsets.append((itemset, support))

            # Recursive call for larger itemsets
            suffix_items = {i for i in items if get_support(prefix.union({i}), transaction_dict) >= min_support}
            eclat(itemset, suffix_items, min_support, transaction_dict, frequent_itemsets)

# Step 4: Run ECLAT algorithm
min_support = 5  # Minimum support count
unique_items = {item for transaction in transaction_dict.values() for item in transaction}

frequent_itemsets = []
eclat(set(), unique_items, min_support, transaction_dict, frequent_itemsets)

# Step 5: Display frequent itemsets
print("Frequent Itemsets:")
for itemset, support in sorted(frequent_itemsets, key=lambda x: -x[1]):
    print(f"Itemset: {itemset}, Support: {support}")
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

dataset_path =  '/content/supermarket_sales - Sheet1.csv'
df = pd.read_csv(dataset_path)


basket = df.groupby(['Invoice ID', 'Product line'])['Product line'].count().unstack().fillna(0)
basket = basket.applymap(lambda x: 1 if x > 0 else 0)


frequent_itemsets = apriori(basket, min_support=0.05, use_colnames=True)


def is_closed(itemset, all_itemsets):
    """Check if an itemset is closed."""
    for other_itemset in all_itemsets:
        if set(itemset).issubset(set(other_itemset)) and itemset != other_itemset:
            return False
    return True


all_itemsets = frequent_itemsets['itemsets'].tolist()
frequent_itemsets['is_closed'] = frequent_itemsets['itemsets'].apply(lambda x: is_closed(x, all_itemsets))

closed_itemsets = frequent_itemsets[frequent_itemsets['is_closed']]


rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7,num_itemsets=0)


print("Closed Itemsets:")
print(closed_itemsets)

print("\nAssociation Rules:")
print(rules)
```

```python
#Pincer search

import pandas as pd
from itertools import combinations
from collections import defaultdict

class PincerSearch:
    def __init__(self, min_support):
        self.min_support = min_support
        self.transactions = []
        self.frequent_itemsets = []
        self.maximal_itemsets = []

    def load_data(self, file_path):
        """Load dataset and prepare transactions."""
        print("Loading data...")
        df = pd.read_csv(file_path)
        self.transactions = df.groupby('Invoice ID')['Product line'].apply(list).tolist()

    def count_itemsets(self, itemsets):
        """Count occurrences of itemsets in transactions."""
        counts = defaultdict(int)
        for transaction in self.transactions:
            for itemset in itemsets:
                if set(itemset).issubset(transaction):
                    counts[itemset] += 1
        return counts

    def generate_candidates(self, itemsets, k):
        """Generate candidate itemsets of size k."""
        items = set(item for itemset in itemsets for item in itemset)
        return [tuple(sorted(comb)) for comb in combinations(items, k)]

    def pincer_search(self):
        """Main Pincer-Search algorithm."""
        # Step 1: Find frequent 1-itemsets
        items = set(item for transaction in self.transactions for item in transaction)
        candidate_1_itemsets = [(item,) for item in items]
        counts = self.count_itemsets(candidate_1_itemsets)
        self.frequent_itemsets = [itemset for itemset, count in counts.items() if count >= self.min_support]

        print(f"Frequent 1-itemsets: {self.frequent_itemsets}")

        k = 2
        while self.frequent_itemsets:
            # Step 2: Forward search (generate candidates of size k)
            candidates = self.generate_candidates(self.frequent_itemsets, k)
            counts = self.count_itemsets(candidates)
            self.frequent_itemsets = [itemset for itemset, count in counts.items() if count >= self.min_support]

            # Step 3: Backward search (prune maximal infrequent supersets)
            self.maximal_itemsets = [
                itemset for itemset in candidates if all(
                    subset not in self.frequent_itemsets for subset in combinations(itemset, k - 1)
                )
            ]

            print(f"Frequent {k}-itemsets: {self.frequent_itemsets}")
            print(f"Maximal frequent itemsets: {self.maximal_itemsets}")

            k += 1

    def run(self, file_path):
        """Run the Pincer-Search algorithm."""
        self.load_data(file_path)
        self.pincer_search()
        print("Final Maximal Frequent Itemsets:")
        for itemset in self.maximal_itemsets:
            print(itemset)

# Example usage
file_path = '/content/supermarket_sales - Sheet1.csv'  # Replace with your dataset path
min_support = 10  # Adjust based on dataset size
ps = PincerSearch(min_support)
ps.run(file_path)
```

```python
import pandas as pd
import hashlib

# Step 1: Load the dataset
# Replace with your dataset's file path
dataset_path = '/content/supermarket_sales - Sheet1.csv'
df = pd.read_csv(dataset_path)

# Step 2: Hash Individual Items
def hash_item(item):
    """Hash an item to a fixed-size hash using SHA256."""
    return hashlib.sha256(item.encode('utf-8')).hexdigest()

# Apply hashing to the 'Item' column
if 'Product line' in df.columns:
    df['Item_Hash'] = df['Product line'].apply(hash_item)

# Step 3: Hash Transactions (if applicable)
# Combine all items in a transaction into a single string and hash
if 'InvoiceNo' in df.columns and 'Item_Hash' in df.columns:
    transaction_hashes = df.groupby('Invoice ID')['Item_Hash'].apply(
        lambda items: hashlib.sha256(''.join(sorted(items)).encode('utf-8')).hexdigest()
    )
    df['Transaction_Hash'] = df['Invoice ID'].map(transaction_hashes)

# Step 4: Save or Display Hashed Data
# Save the hashed dataset to a new file
hashed_dataset_path = 'hashed_supermarket_sales.csv'
df.to_csv(hashed_dataset_path, index=False)

print("Sample of Hashed Dataset:")
print(df.head())
```

```python
import numpy as np
import pandas as pd

# Define entropy and information gain calculation
def entropy(data):
    _, counts = np.unique(data[:, -1], return_counts=True) # Use standard indexing for NumPy array
    prob = counts / len(data)
    return -np.sum(prob * np.log2(prob))

def information_gain(data, feature):
    total_entropy = entropy(data)
    values, counts = np.unique(data[:, feature], return_counts=True) # Use standard indexing for NumPy array
    weighted_entropy = np.sum((counts / len(data)) * [entropy(data[data[:, feature] == v]) for v in values]) # Use standard indexing for NumPy array
    return total_entropy - weighted_entropy

# Define the ID3 algorithm
def id3(data, features):
    if len(np.unique(data[:, -1])) == 1: # Use standard indexing for NumPy array
        return np.unique(data[:, -1])[0]

    if len(features) == 0:
        return np.unique(data[:, -1])[0] # Use standard indexing for NumPy array

    # Select the feature with the maximum information gain
    gains = [information_gain(data, feature) for feature in features]
    best_feature = features[np.argmax(gains)]

    tree = {best_feature: {}}
    features = [f for f in features if f != best_feature]

    for value in np.unique(data[:, best_feature]):  # Use standard indexing for NumPy array
        subtree = id3(data[data[:, best_feature] == value], features) # Use standard indexing for NumPy array
        tree[best_feature][value] = subtree

    return tree

# Example usage with some dataset
data =  pd.read_csv('/content/supermarket_sales - Sheet1.csv')

# Assuming your 'features' are column indices
features = [0, 1, 2, 3]  # Example: Indices of your features

# Convert the relevant columns to a NumPy array for ID3
data_for_id3 = data[['Gender', 'Unit price', 'Quantity', 'Tax 5%', 'Customer type']].values  # Include target column

# Build the tree
tree = id3(data_for_id3, features)
print(tree)
```

```python
#divisive clustering(hierarchical method )
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from scipy.spatial.distance import cdist
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt



print(data.head())

# Select numerical columns or drop non-numeric columns (adjust based on your dataset)
numerical_data = data[['Unit price', 'Quantity', 'Total']]  # Modify based on your dataset

# Standardize the data
scaler = StandardScaler()
numerical_data_scaled = scaler.fit_transform(numerical_data)

# Function for Divisive Hierarchical Clustering
def divisive_hierarchical_clustering(X, n_clusters):
    clusters = [X]

    while len(clusters) < n_clusters:
        largest_cluster = max(clusters, key=lambda cluster: len(cluster))

        kmeans = KMeans(n_clusters=2, random_state=42)
        kmeans.fit(largest_cluster)

        cluster_1 = largest_cluster[kmeans.labels_ == 0]
        cluster_2 = largest_cluster[kmeans.labels_ == 1]

        clusters.remove(largest_cluster)
        clusters.append(cluster_1)
        clusters.append(cluster_2)

    return clusters

# Apply Divisive Hierarchical Clustering
n_clusters = 4  # Adjust as necessary
clusters = divisive_hierarchical_clustering(numerical_data_scaled, n_clusters)

# Visualize the results (Dendrogram)
Z = linkage(numerical_data_scaled, method='ward')

plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()  # Ensure this is called to display the plot

# Assign clusters to the data using fcluster
labels = fcluster(Z, n_clusters, criterion='maxclust')
data['Cluster'] = labels

# Visualize the dataset with the assigned clusters
plt.figure(figsize=(8, 6))
plt.scatter(data['Quantity'], data['Total'], c=data['Cluster'], cmap='viridis')
plt.title('Clusters based on Divisive Hierarchical Clustering')
plt.xlabel('Quantity')
plt.ylabel('Total')
plt.show()  # Ensure this is called to display the plot

# Optionally, you can inspect the clusters
for i in range(1, n_clusters + 1):
    print(f"Cluster {i} Summary:")
    print(data[data['Cluster'] == i].describe())
```

```python
#ai dbscan
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/content/supermarket_sales - Sheet1.csv')  # Replace with your file path

# Select numerical features for clustering (e.g., Total, Quantity, Unit price)
features = df[['Total', 'Quantity', 'Unit price']]

# Scale the features (DBSCAN is sensitive to feature scales)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)  # Adjust eps and min_samples based on dataset
df['Cluster'] = dbscan.fit_predict(scaled_features)

# Count clusters and noise points
n_clusters = len(set(df['Cluster'])) - (1 if -1 in df['Cluster'].values else 0)
n_noise = list(df['Cluster']).count(-1)

print(f"Number of clusters: {n_clusters}")
print(f"Number of noise points: {n_noise}")

# Display the cluster distribution
print("\nCluster distribution:")
print(df['Cluster'].value_counts())

# Visualize clusters (e.g., using Total and Unit price)
plt.scatter(df['Total'], df['Unit price'], c=df['Cluster'], cmap='viridis', s=50)
plt.xlabel('Total')
plt.ylabel('Unit price')
plt.title('DBSCAN Clustering')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# Drop rows with null values
df = df.na.drop()

# Add a profit column (example calculation: profit = Total - (Unit Price * Quantity))
from pyspark.sql.functions import col
df = df.withColumn("Profit", col("Total") - (col("Unit Price") * col("Quantity")))

# Show updated data
df.show(5)
```

```python
#FIM
from pyspark.sql import functions as F

# Combine product names into a list for each transaction
transactions = df.groupBy("Invoice ID").agg(F.collect_list("Product line").alias("items"))

# Show the prepared dataset
transactions.show(truncate=False)
```

```python
from pyspark.ml.fpm import FPGrowth

# Create FP-Growth model
fpGrowth = FPGrowth(itemsCol="items", minSupport=0.05, minConfidence=0.3)
model = fpGrowth.fit(transactions)

# Display frequent itemsets
print("Frequent Itemsets:")
model.freqItemsets.show()

# Display association rules
print("Association Rules:")
model.associationRules.show()

# Transform data to show predictions (optional)
predictions = model.transform(transactions)
predictions.show(truncate=False)
```

```python
from itertools import combinations

# Define minimum support
min_support = 0.2
num_transactions = transactions.count()

# Step 1: Count individual items
items_counts = transactions.withColumn("item", F.explode("items")) \
    .groupBy("item") \
    .count() \
    .filter(F.col("count") / num_transactions >= min_support)

frequent_items = items_counts.select("item").rdd.flatMap(lambda x: x).collect()

# Step 2: Generate candidate itemsets and count support iteratively
frequent_itemsets = []
k = 2
current_frequent_items = frequent_items

while current_frequent_items:
    # Generate candidate itemsets
    candidates = list(combinations(current_frequent_items, k))

    # Count support for candidates
    candidates_df = spark.createDataFrame(candidates, schema=["itemset"])
    transactions_with_candidates = transactions.crossJoin(candidates_df) \
        .withColumn("contains", F.array_contains("items", F.col("itemset")))

    candidate_counts = transactions_with_candidates \
        .filter("contains") \
        .groupBy("itemset") \
        .count() \
        .filter(F.col("count") / num_transactions >= min_support)

    # Collect frequent itemsets
    current_frequent_items = candidate_counts.select("itemset").rdd.flatMap(lambda x: x).collect()
    frequent_itemsets.extend(current_frequent_items)
    k += 1
```

```python
from pyspark.ml.feature import StringIndexer, VectorAssembler
from pyspark.ml.classification import DecisionTreeClassifier

# Check if 'label' column already exists
if 'label' in df.columns:
    df = df.drop('label')  # Drop the existing 'label' column

# Now, apply StringIndexer to create a new 'label' column
indexer = StringIndexer(inputCol="Customer type", outputCol="label")
df = indexer.fit(df).transform(df)

# Assemble features
assembler = VectorAssembler(
    inputCols=["Unit price", "Quantity", "Total", "Rating"], outputCol="features"
)
df = assembler.transform(df)

# Split data
training_data, test_data = df.randomSplit([0.8, 0.2], seed=42)

# Train Decision Tree Classifier
dt = DecisionTreeClassifier(featuresCol="features", labelCol="label")
model = dt.fit(training_data)

# Make predictions
predictions = model.transform(test_data)
predictions.select("Customer type", "features", "prediction").show(5)
```

```python
from pyspark.ml.classification import NaiveBayes

# Create a Naive Bayes classifier
nb = NaiveBayes(featuresCol="features", labelCol="label", modelType="multinomial")

# Train the model
nb_model = nb.fit(training_data)
```

```python
# Make predictions on the test data
predictions = nb_model.transform(test_data)

# Show predictions
predictions.select("features", "label", "prediction").show(10)
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

# Evaluate model accuracy
evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator.evaluate(predictions)
print(f"Accuracy: {accuracy:.2f}")
```

```python
from pyspark.ml.clustering import KMeans

# Use KMeans clustering
kmeans = KMeans(featuresCol="features", k=3)
model = kmeans.fit(df)

# Get cluster predictions
clusters = model.transform(df)
clusters.select("City", "features", "prediction").show(5)
```

```python
from pyspark.ml.regression import LinearRegression

# Train Linear Regression model
lr = LinearRegression(featuresCol="features", labelCol="Total")
lr_model = lr.fit(training_data)

# Make predictions
sales_predictions = lr_model.transform(test_data)
sales_predictions.select("features", "Total", "prediction").show(5)
```

```python
from pyspark.ml.feature import VectorAssembler

# Check if 'features' column already exists
if 'features' not in df.columns:
    # Assemble features into a vector only if it doesn't exist
    assembler = VectorAssembler(inputCols=["Unit price", "Quantity", "Total", "Rating"], outputCol="features")
    data_with_features = assembler.transform(df)
else:
    # If 'features' column exists, use the existing DataFrame
    data_with_features = df

# Convert to Pandas for Scikit-learn compatibility
data_pandas = data_with_features.select("features").toPandas()

# Extract features as a NumPy array
import numpy as np
X = np.array(data_pandas["features"].tolist())
```

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Apply DBSCAN
dbscan = DBSCAN(eps=10, min_samples=5)  # Adjust `eps` and `min_samples` based on your data
labels = dbscan.fit_predict(X)

# Add labels to the dataset
data_pandas["Cluster"] = labels
```
